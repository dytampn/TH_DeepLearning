```python
# ==========================================
# CNN Cat vs Dog using CIFAR10
# ==========================================

import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np

# ==========================================
# 1. Load CIFAR10
# ==========================================

(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()

# ==========================================
# 2. Lọc dữ liệu chỉ lấy Cat và Dog
# Cat = 3
# Dog = 5
# ==========================================

cat_label = 3
dog_label = 5

train_filter = np.where(
    (y_train == cat_label) |
    (y_train == dog_label)
)[0]

test_filter = np.where(
    (y_test == cat_label) |
    (y_test == dog_label)
)[0]

x_train = x_train[train_filter]
y_train = y_train[train_filter]

x_test = x_test[test_filter]
y_test = y_test[test_filter]

# ==========================================
# 3. Chuyển nhãn
# Cat = 0
# Dog = 1
# ==========================================

y_train = np.where(y_train == dog_label, 1, 0)
y_test = np.where(y_test == dog_label, 1, 0)

print("Train shape:", x_train.shape)
print("Test shape:", x_test.shape)

# ==========================================
# 4. Chuẩn hóa dữ liệu
# ==========================================

x_train = x_train / 255.0
x_test = x_test / 255.0

# ==========================================
# 5. Hiển thị ảnh mẫu
# ==========================================

class_names = ['Cat', 'Dog']

plt.figure(figsize=(10, 5))

for i in range(10):
    plt.subplot(2, 5, i + 1)

    plt.imshow(x_train[i])

    label = int(y_train[i])
    plt.title(class_names[label])

    plt.axis("off")

plt.tight_layout()
plt.show()

# ==========================================
# 6. Xây dựng mô hình CNN
# ==========================================

model = models.Sequential([

    layers.Conv2D(
        32,
        (3, 3),
        activation='relu',
        input_shape=(32, 32, 3)
    ),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(
        64,
        (3, 3),
        activation='relu'
    ),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(
        128,
        (3, 3),
        activation='relu'
    ),

    layers.Flatten(),

    layers.Dense(
        128,
        activation='relu'
    ),

    layers.Dropout(0.3),

    # Output nhị phân
    layers.Dense(
        1,
        activation='sigmoid'
    )
])

# ==========================================
# 7. Compile model
# ==========================================

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()

# ==========================================
# 8. Train model
# ==========================================

history = model.fit(
    x_train,
    y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.2
)

# ==========================================
# 9. Đánh giá mô hình
# ==========================================

test_loss, test_accuracy = model.evaluate(
    x_test,
    y_test
)

print("\nTest Accuracy:", test_accuracy)
print("Test Loss:", test_loss)

# ==========================================
# 10. Vẽ biểu đồ Accuracy và Loss
# ==========================================

plt.figure(figsize=(12, 5))

# Accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'])

# Loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Validation'])

plt.tight_layout()
plt.show()

# ==========================================
# 11. Predict ảnh test
# ==========================================

predictions = model.predict(x_test)

plt.figure(figsize=(12, 6))

for i in range(10):

    plt.subplot(2, 5, i + 1)

    plt.imshow(x_test[i])

    pred = predictions[i][0]

    predicted_label = 1 if pred >= 0.5 else 0
    true_label = int(y_test[i])

    plt.title(
        f"Pred: {class_names[predicted_label]}\n"
        f"True: {class_names[true_label]}"
    )

    plt.axis('off')

plt.tight_layout()
plt.show()

# ==========================================
# 12. Lưu model
# ==========================================

model.save("cnn_cat_dog_cifar10.h5")

print("\nĐã lưu model: cnn_cat_dog_cifar10.h5")
```

```python
# ==========================================
# Upload ảnh và nhận diện Cat/Dog
# ==========================================

import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files # Import files from google.colab
import io # Import io for BytesIO

# ==========================================
# 1. Load model đã train
# ==========================================

model = load_model("cnn_cat_dog_cifar10.h5")

print("Đã load model thành công!")

# ==========================================
# 2. Upload ảnh từ máy
# ==========================================

uploaded = files.upload() # Use google.colab.files.upload()

# Check if any file was uploaded
if not uploaded:
    print("Bạn chưa chọn ảnh hoặc quá trình tải lên thất bại!")
    exit()

# Get the first uploaded file's name and content
file_name = next(iter(uploaded))
file_content = uploaded[file_name]

print("Ảnh đã chọn:", file_name)

# ==========================================
# 3. Đọc và tiền xử lý ảnh
# ==========================================

# Load image from bytes content
img = image.load_img(
    io.BytesIO(file_content),
    target_size=(32, 32)  # CIFAR10 size
)

img_array = image.img_to_array(img)

# Chuẩn hóa
img_array = img_array / 255.0

# Thêm chiều batch
img_array = np.expand_dims(img_array, axis=0)

# ==========================================
# 4. Predict
# ==========================================

prediction = model.predict(img_array)[0][0]

class_names = ['Cat', 'Dog']

predicted_label = 1 if prediction >= 0.5 else 0

confidence = (
    prediction if predicted_label == 1
    else (1 - prediction)
)

# ==========================================
# 5. Hiển thị kết quả
# ==========================================

# Re-load image for display (matplotlib can directly plot PIL Image objects)
img_show = image.load_img(io.BytesIO(file_content))

plt.figure(figsize=(6, 6))
plt.imshow(img_show)

plt.title(
    f"Prediction: {class_names[predicted_label]}\n"
    f"Confidence: {confidence*100:.2f}%"
)

plt.axis('off')
plt.show()

# ==========================================
# 6. In kết quả
# ==========================================

print("\n===== KẾT QUẢ NHẬN DIỆN =====")
print("Dự đoán:", class_names[predicted_label])
print(f"Độ tin cậy: {confidence*100:.2f}%")
```
